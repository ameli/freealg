```python
import numpy
from freealg.distributions import MarchenkoPastur
from freealg import FreeForm
latex = True
```

```python
mp = MarchenkoPastur(1/50)
A = mp.matrix(3000)
eig = numpy.linalg.eigvalsh(A)
```

```python
ff = FreeForm(eig, support=(mp.lam_m, mp.lam_p))
psi = ff.fit(method='jacobi', K=10, alpha=0.5, beta=0.5, reg=0, projection='gaussian',
             kernel_bw=0.1, damp='jackson', force=False, continuation='pade', 
             pade_p=1, pade_q=1, odd_side='left', pade_reg=0, optimizer='ls',
             plot=False, latex=latex)
```

```python
x = numpy.linspace(-0.5, 2.5, 300)

# Aspect tario for NeurIPS logo upload (320 width, 256 height)
aspect_ratio = 256.0 / 320.0
y_range = aspect_ratio * (x[-1] - x[0])

y = numpy.linspace(-y_range / 2.0, y_range / 2.0, 250)

m1, m2 = ff.stieltjes(x, y, plot=True, latex=latex, save='logo.svg')
```
