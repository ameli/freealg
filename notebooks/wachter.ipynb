```python
import numpy
import scipy
from freealg.distributions import Wachter
wa = Wachter(2, 3)

latex = True
ext = '.png'
```

```python
x = numpy.linspace(-0.2, 1, 500)
rho = wa.density(x, plot=True, latex=latex, save='wa_density' + ext)
```

```python
s = wa.sample(1000, plot=True, latex=latex, save='wa_samples' + ext)
```

```python
x = numpy.linspace(wa.lam_m-1, wa.lam_p+1, 400)
hilb = wa.hilbert(x, plot=True, latex=latex, save='wa_hilbert' + ext)
```

```python
x = numpy.linspace(wa.lam_m-1, wa.lam_p+1, 400)
y = numpy.linspace(-1, 1, 300)
m1, m2 = wa.stieltjes(x, y, plot=True, latex=latex, save='wa_stieltjes' + ext)
```

```python
m1_D, m2_D = wa.stieltjes(plot=True, on_disk=True, latex=latex, save='wa_stieltjes_disk' + ext)
```
