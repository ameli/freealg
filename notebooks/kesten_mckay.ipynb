```python
import numpy
from freealg.distributions import KestenMcKay
km = KestenMcKay(3)

latex = True
ext = '.png'
```

```python
rho = km.density(plot=True, latex=latex, save='km_density' + ext)
```

```python
s = km.sample(1000, plot=True, latex=latex, save='km_samples' + ext)
```

```python
x = numpy.linspace(km.lam_m-5, km.lam_p+5, 400)
hilb = km.hilbert(x, plot=True, latex=latex, save='km_hilbert' + ext)
```

```python
x = numpy.linspace(km.lam_m-3, km.lam_p+3, 400)
y = numpy.linspace(-3, 3, 300)
m1, m2 = km.stieltjes(x, y, plot=True, latex=latex, save='km_stieltjes' + ext)
```

```python
m1_D, m2_D = km.stieltjes(plot=True, on_disk=True, latex=latex, save='km_stieltjes_disk' + ext)
```
