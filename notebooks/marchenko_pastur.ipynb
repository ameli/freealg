```python
from freealg.distributions import MarchenkoPastur
mp = MarchenkoPastur(1/50)

latex = True
ext = '.png'
```

```python
rho = mp.density(plot=True, latex=latex, save='mp_density' + ext)
```

```python
s = mp.sample(1000, plot=True, latex=latex, save='mp_samples' + ext)
```

```python
hilb = mp.hilbert(plot=True, latex=latex, save='mp_hilbert' + ext)
```

```python
m1, m2 = mp.stieltjes(plot=True, latex=latex, save='mp_stieltjes' + ext)
```

```python
m1_D, m2_D = mp.stieltjes(plot=True, on_disk=True, latex=latex, save='mp_stieltjes_disk' + ext)
```
