```python
import numpy
from freealg.distributions import Wigner
wg = Wigner()

latex = True
ext = '.png'
```

```python
rho = wg.density(plot=True, latex=latex, save='wg_density' + ext)
```

```python
s = wg.sample(1000, plot=True, latex=latex, save='wg_samples' + ext)
```

```python
x = numpy.linspace(wg.lam_m-5, wg.lam_p+5, 400)
hilb = wg.hilbert(x, plot=True, latex=latex, save='wg_hilbert' + ext)
```

```python
x = numpy.linspace(wg.lam_m-3, wg.lam_p+3, 400)
y = numpy.linspace(-3, 3, 300)
m1, m2 = wg.stieltjes(x, y, plot=True, latex=latex, save='wg_stieltjes' + ext)
```

```python
m1_D, m2_D = wg.stieltjes(plot=True, on_disk=True, latex=latex, save='wg_stieltjes_disk' + ext)
```
