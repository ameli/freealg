```python
import numpy
import scipy
from freealg.distributions import Meixner
mx = Meixner(1, 0.2)

latex = True
ext = '.png'
```

```python
x = numpy.linspace(-3, 6, 500)
rho = mx.density(x, plot=True, latex=latex, save='mx_density' + ext)
```

```python
s = mx.sample(1000, plot=True, latex=latex, save='mx_samples' + ext)
```

```python
x = numpy.linspace(mx.lam_m-1, mx.lam_p+1, 400)
hilb = mx.hilbert(x, plot=True, latex=latex, save='mx_hilbert' + ext)
```

```python
x = numpy.linspace(mx.lam_m-4, mx.lam_p+4, 400)
y = numpy.linspace(-4, 4, 300)
m1, m2 = mx.stieltjes(x, y, plot=True, latex=latex, save='mx_stieltjes' + ext)
```

```python
m1_D, m2_D = mx.stieltjes(plot=True, on_disk=True, latex=latex, save='mx_stieltjes_disk' + ext)
```
